__g)__ Increase the price of the order lines by 10% for those products whose price is greater than 80 Euros. To do this, create a new column called `Descuento`.

* Create a function that receives as a parameter an integer value that represents the price. If this value is greater than 80, the function returns it increased by 10%. If the value is less than or equal to 80, the function returns that value unchanged.

* Apply the above function using `map` on column `precio`.

```python
# Sol:
def discount(price: int):
  return round((price, price * 1.1) [price > 80], 2)

pedidos['Descuento'] = pedidos.precio.map(discount)
```
